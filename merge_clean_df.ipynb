```python
import pandas as pd
import sys
import numpy as np 
```

```python
# Read the CSV file
df = pd.read_csv('merged_keywords_연금_상해_cleaned.csv', encoding='utf-8')
# Check if the DataFrame is empty
print(df.isnull().sum(), "\n",df.shape)
```

```python
df.head()
```

```python
df['date'] = df['date'].str.split(' ').str[0]           # '2023.08.18.'만 추출
df['date'] = df['date'].str.replace('.', '-', regex=False)  # 점을 -로 변경
df['date'] = df['date'].str.rstrip('-')                 # 마지막 '-' 제거
```

```python
df['date']
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
df = df.sort_values('date').reset_index(drop=True)
```

```python
df
```

```python
df.to_csv('merged_keywords_연금_상해_cleaned_sorted.csv', index=False, encoding='utf-8-sig')
```
